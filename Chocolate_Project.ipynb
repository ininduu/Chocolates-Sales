## Cleaning Data

```python
# Ubah ke dalam Float karena kolom masih string
choco['Amount'] = choco['Amount'].replace('[\$,]', '', regex=True)
choco['Amount'] = choco['Amount'].astype(float)

# Ubah ke format datetime
choco['Date'] = pd.to_datetime(choco["Date"])

# Simpan ke dalam variabel baru
choco_clean = choco
choco_clean.head()
```

```python
choco_clean.info()
```

```python
# Simpan ke dalam Excel
sales_chocolates = choco_clean.to_excel('sales_chocolates.xlsx', index=False)
print(sales_chocolates)
```

```python
product_sales = choco_clean.groupby('Product')['Amount'].sum().reset_index()
product_sales = product_sales.sort_values(by='Amount', ascending=False)
display(product_sales)
```
